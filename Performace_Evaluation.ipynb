### generate code for training decisin tree and evaluate with 10-fold cross validation

```python

from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Load the Iris dataset
iris = load_iris()

# Split the dataset into features (X) and target (y)
X = iris.data
y = iris.target

# Create an instance of the DecisionTreeClassifier class
clf = DecisionTreeClassifier()

# Perform 10-fold cross-validation and compute the mean accuracy
scores = cross_val_score(clf, X, y, cv=10)
mean_accuracy = scores.mean()

# Print the mean accuracy
print("Mean accuracy:", mean_accuracy)
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import LeaveOneOut, cross_val_score

# Load the iris dataset
iris = load_iris()

# Split the data into features (X) and labels (y)
X = iris.data
y = iris.target

# Create an instance of the decision tree classifier
clf = DecisionTreeClassifier()

# Create a leave-one-out cross-validation object
loo = LeaveOneOut()

# Evaluate the decision tree classifier using leave-one-out cross-validation
scores = cross_val_score(clf, X, y, cv=loo)

# Compute the mean accuracy and standard deviation and print them out
mean_accuracy = scores.mean()
std_dev = scores.std()
print("Mean accuracy:", mean_accuracy)
print("Standard deviation:", std_dev)
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
## Define feature names
feature_names = iris_data.feature_names

# Define target names
target_names = iris_data.target_names

# Define Decision Tree Classifier
clf = DecisionTreeClassifier()

# Define KFold with 10 folds
kf = KFold(n_splits=10, shuffle=True, random_state=6)

# Initialize lists to store results
accuracies = []
feature_importances = []

# Loop through each fold
for fold, (train_index, test_index) in enumerate(kf.split(iris_data.data)):
    
    # Get train and test data
    X_train, X_test = iris_data.data[train_index], iris_data.data[test_index]
    y_train, y_test = iris_data.target[train_index], iris_data.target[test_index]
    
    # Fit classifier on training data
    clf.fit(X_train, y_train)
    
    # Make predictions on test data
    y_pred = clf.predict(X_test)
    
    # Calculate accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store accuracy score in list
    accuracies.append(accuracy)
    
    # Get feature importances
    feature_importance = clf.feature_importances_
    
    # Store feature importances in list
    feature_importances.append(feature_importance)

    # Create a DataFrame to display the iris data for each fold
    iris_data1 = np.concatenate((X_test, y_test.reshape(-1,1)), axis=1)
    columns = iris_data['feature_names']
    X_df = pd.DataFrame(data=iris_data1, columns=columns+['label'])
    X_df['label'] = y_test
    df = X_df.sample(frac=1, random_state=6)
    df['fold'] = fold+1

    # Print fold number and accuracy score
    print(f"Fold {fold+1}")
    print("Cross-validation accuracy: %0.2f (+/- %0.2f)" % (np.mean(accuracies), np.std(accuracies) * 2))
    
    # Show sample of data for current fold
    print(df)
    
    print("\n")
```
