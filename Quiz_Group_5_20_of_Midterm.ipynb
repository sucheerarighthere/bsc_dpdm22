## รวมตารางความเร็วลม ทั้ง 5 จังหวัด

```python
all_province_wind = pd.concat([winds_lei, winds_skn, winds_npm, winds_nk, winds_kn])
```

```python
wind = pd.melt(all_province_wind,
        id_vars=['จังหวัด','สถานี','เดือน','ปี'],
                 var_name='วันที่', value_name='ความเร็วลม')
```

```python
wind['วันที่'] = wind['วันที่'].astype(int)
```

```python
wind = wind.sort_values(by=['จังหวัด','ปี','เดือน','วันที่']).reset_index().iloc[:,1:]
```
